#### IMPORTS

```python
from pprint import pprint

import torch
import matplotlib.pyplot as plt

from utils.fsys import list_files
```

### NETWORK INSTANTIATION

```python
file = sorted(list_files('resources/models/onelayer_sanitization/*.pt'))[-1]
stats = torch.load(file)['stats_mem']

pprint(stats)
```

```python
ta = [s["t_accuracy"] for s in stats]
va = [s["v_accuracy"] for s in stats]
tl = [s["t_loss"] for s in stats]
vl = [s["v_loss"] for s in stats]
```

### PLOTS

```python
# PLOTS
plt.figure(figsize=(20,10))
plt.ylim([-.05, 1.05])
plt.grid()
plt.plot(tl)
plt.plot(vl)
plt.plot(ta)
plt.plot(va)
plt.legend(['train loss', 'validation loss', 'train accuracy', 'validation accuracy'])
plt.show()
```

### BEST MODEL

```python
print(f'TA={ta[-1]:.5f}, VA={va[-1]:.5f}')
print(f'best VL={min(vl):.5f} ({vl.index(min(vl))})')
print(f'     VA={va[vl.index(min(vl))]:.5f}')
```
